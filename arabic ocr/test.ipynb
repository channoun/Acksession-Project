```python
import os
import cv2
import matplotlib.pyplot as plt


for image_path in os.listdir("./arabic_compiled_dataset/test/images/"):
    image = cv2.imread("./arabic_compiled_dataset/test/images/" + image_path)
    label_path = "./arabic_compiled_dataset/test/labels/" + image_path[:-3] + "txt"
    dh, dw, _ = image.shape
    with open(label_path, "r") as f:
        write = False
        lines = f.readlines()
        for line in lines:
            class_id, x, y, w, h = map(float, line.strip().split(" "))
            print(image_path)
            l = int((x - w / 2) * dw)
            r = int((x + w / 2) * dw)
            t = int((y - h / 2) * dh)
            b = int((y + h / 2) * dh)
    
            if l < 0:
                l = 0
            if r > dw - 1:
                r = dw - 1
            if t < 0:
                t = 0
            if b > dh - 1:
                b = dh - 1

            cv2.rectangle(image, (l, t), (r, b), (0, 0, 255), 1)
    if not os.path.exists("./boxes"):
        os.makedirs("./boxes")
    cv2.imwrite(f"./boxes/{image_path}", image)
```
